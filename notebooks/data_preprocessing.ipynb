```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Correct paths (based on your structure)
real_path = "data/train/REAL"
fake_path = "data/train/FAKE"

# Count images
print("Real Images:", len(os.listdir(real_path)))
print("Fake Images:", len(os.listdir(fake_path)))
```

```python
# Show sample images
def show_images(folder, label):
    images = os.listdir(folder)[:5]

    plt.figure(figsize=(10,5))
    for i, img_name in enumerate(images):
        img_path = os.path.join(folder, img_name)
        img = Image.open(img_path)

        plt.subplot(1,5,i+1)
        plt.imshow(img)
        plt.title(label)
        plt.axis("off")

    plt.show()

show_images(real_path, "Real")
show_images(fake_path, "Fake")
```

```python
# Check image sizes
sizes = []

for folder in [real_path, fake_path]:
    for img_name in os.listdir(folder)[:50]:
        img_path = os.path.join(folder, img_name)
        img = Image.open(img_path)
        sizes.append(img.size)

print("Sample sizes:", sizes[:10])
```
